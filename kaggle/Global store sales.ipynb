```python
# Import libraries
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Load the Global Superstore dataset
df = pd.read_csv('C:\\Users\ESHOP\\Documents\\Kaggle Practice Project\\Datasets\\Global_Superstore2.csv\\Global_Superstore2.csv', encoding='latin1') 
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Data preprocessing
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Function to display KPIs
def display_kpis(filtered_df):
    total_sales = filtered_df['Sales'].sum()
    total_profit = filtered_df['Profit'].sum()
    num_orders = filtered_df['Order ID'].nunique()
    avg_order_value = total_sales / num_orders if num_orders > 0 else 0
    
    print(f'Total Sales: ${total_sales:,.2f}')
    print(f'Total Profit: ${total_profit:,.2f}')
    print(f'Number of Orders: {num_orders}')
    print(f'Average Order Value: ${avg_order_value:,.2f}')
```

```python
region_dropdown = widgets.Dropdown(
    options=['All'] + df['Region'].unique().tolist(),
    value='All',
    description='Select Region:',
)
```

```python
# Function to update KPIs and visualizations based on selected region
def update_dashboard(region):
    if region != 'All':
        filtered_df = df[df['Region'] == region]
    else:
        filtered_df = df

    # Display KPIs
    display_kpis(filtered_df)

    # Sales by Category
    category_sales = filtered_df.groupby('Category')['Sales'].sum().reset_index()
    fig_category = px.bar(category_sales, x='Category', y='Sales', title='Sales by Category')
    fig_category.show()

    # Sales over Time
    sales_over_time = filtered_df.groupby(filtered_df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
    sales_over_time['Order Date'] = sales_over_time['Order Date'].dt.to_timestamp()
    
    fig_time = px.line(sales_over_time, x='Order Date', y='Sales', title='Sales Over Time')
    fig_time.show()

    # Scatter Plot for Sales by Country
    fig_map = px.scatter_geo(filtered_df, 
                             locations="Country", 
                             size="Sales", 
                             hover_name="City",
                             title='Sales by Country',
                             projection="natural earth")
    fig_map.show()

# Display the dropdown and link it to the update function
widgets.interactive(update_dashboard, region=region_dropdown)
```
